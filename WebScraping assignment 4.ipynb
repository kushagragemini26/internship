Scrape the details most watched tv series of all time from imdb.com. 
Url = https://www.imdb.com/list/ls095964455/ You have 
to find the following details: 
A) Name 
B) Year span 
C) Genre 
D) Run time 
E) Ratings 
F) Votes

```python
driver = webdriver.Chrome()
```

```python
driver.get('https://www.imdb.com/list/ls095964455/')
```

```python
Name=[]
Year_Span=[] 
Genre=[] 
Run_Time=[] 
Ratings=[] 
Votes=[]
```

```python
# Let's create a function
name = driver.find_elements(By.XPATH, '//h3[@class="lister-item-header"]/a')
for i in name:
    Name.append(i.text)
year = driver.find_elements(By.XPATH, '//span[@class="lister-item-year text-muted unbold"]')
for i in year:
    Year_Span.append(i.text)
gen = driver.find_elements(By.XPATH, '//span[@class="genre"]')
for i in gen:
    Genre.append(i.text)
runtime = driver.find_elements(By.XPATH, '//span[@class="runtime"]')
for i in runtime:
    Run_Time.append(i.text)
ratings = driver.find_elements(By.XPATH, '//div[@class="ipl-rating-star small"]/span[2]')
for i in ratings:
    Ratings.append(i.text)
vote = driver.find_elements(By.XPATH, '//span[@name="nv"]')
for i in vote:
    Votes.append(i.text)
```

```python
# Let's create the dataframe from the scraped data
IMDB=pd.DataFrame({})
IMDB['Name']=Name
IMDB['Year Span']= Year_Span
IMDB['Genre']=Genre
IMDB['Run Time']= Run_Time
IMDB['Ratings']=Ratings
IMDB['Votes']=Votes
IMDB
```
